```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import HillClimbSearch, TreeSearch, K2Score, BicScore, BDeuScore, BayesianEstimator, MaximumLikelihoodEstimator
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('data/hourly/hourly_data.csv')
stations = train.iloc[:, 4:]
stations = stations.fillna(0).astype(int)
train, test = train_test_split(stations, test_size=0.2,random_state=50, shuffle=True)
```

## Structure Learning

```python
est = HillClimbSearch(train)
best_model = est.estimate(scoring_method = K2Score(train))
```

```python
import networkx as nx
nx.draw_circular(best_model, with_labels = True)
```

## Estimate Parameters

```python
model = BayesianNetwork(best_model.edges())
```

```python
model.fit(train, estimator=BayesianEstimator, prior_type = 'K2')
```

```python
model.check_model()
```

```python
train_acc_dict = {}
for col in stations.columns:
    inputs = train.drop(columns=col)
    pred = model.predict(inputs)
    true_val = train[col].values
    acc = (true_val == pred.values).mean()
    train_acc_dict[col] = acc
```

```python
train_acc_dict
```

```python
test_acc_dict = {}
for col in stations.columns:
    inputs = test.drop(columns=col)
    pred = model.predict(inputs)
    true_val = test[col].values
    acc = (true_val == pred.values).mean()
    test_acc_dict[col] = acc
```

```python
test_acc_dict
```

```python
train_k2 = K2Score(train)
test_k2 = K2Score(test)
print(train_k2.score(model))
print(test_k2.score(model))
```

#### Try new ordering

The pollution may travel with wind. We can try a new order of nodes to see if it improves.

```python
stations
```

```python
# Ordering from South-West to North-East
# See if the K2 score improves
# 1 - 4 - 10 - 8 - 5 - 7 - 12 - 6 - 9 - 2 - 3 - 11
new_order = [0, 8, 2, 4, 7, 5, 9, 6, 3, 11, 10, 1]
new_order_stations = stations.iloc[:, new_order]
train_ordered, test_ordered = train_test_split(new_order_stations, test_size=0.2, shuffle = True, random_state=50)
```

```python
new_est = HillClimbSearch(train_ordered)
new_best_model = new_est.estimate(scoring_method=K2Score(train_ordered))
```

```python
nx.draw_circular(new_best_model, with_labels = True)
```

```python
# Fitting new parameters
model2 = BayesianNetwork(new_best_model.edges())
model2.fit(train_ordered, estimator=BayesianEstimator, prior_type = 'K2')
```

```python
train_acc_dict_ord = {}
for col in stations.columns:
    inputs = train_ordered.drop(columns=col)
    pred = model2.predict(inputs)
    true_val = train_ordered[col].values
    acc = (true_val == pred.values).mean()
    train_acc_dict_ord[col] = acc
```

```python
test_acc_dict_ord = {}
for col in stations.columns:
    inputs = test_ordered.drop(columns=col)
    pred = model2.predict(inputs)
    true_val = test_ordered[col].values
    acc = (true_val == pred.values).mean()
    test_acc_dict_ord[col] = acc
```

```python
test_acc_dict_ord
```

```python
train_k2 = K2Score(train)
test_k2 = K2Score(test)
print(train_k2.score(model2))
print(test_k2.score(model2))
```
